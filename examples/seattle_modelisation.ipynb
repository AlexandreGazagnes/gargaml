# 3. Modelisation

```python
## data split
```

```python
df
```

```python
data = Data.dataclass(df, "siteenergyuse_kbtu", test_size=0.33)
data
```

```python
data.X
```

```python
data.y
```

```python
data.train.X
```

```python
data.train.y
```

```python
token = secrets.token_hex(4)
display(token)

pipe = Pipeline(
    [
        ("imputer", KNNImputer()),
        ("scaler", StandardScaler()),
        ("estimator", RandomForestClassifier()),
    ]
)

param_grid = {
    "scaler": [
        StandardScaler(),
        # QuantileTransformer(n_quantiles=100),
        # Normalizer(),
        "passthrough",
    ],
    "estimator": [
        DummyRegressor(),
        # RandomForestRegressor(),
        # XGBRegressor(),
        LinearRegression(),
    ],
}

grid = GridSearchCV(
    pipe,
    param_grid=param_grid,
    cv=3,
    n_jobs=1,
    verbose=1,
    return_train_score=True,
)

grid
```

```python
grid.fit(data.X.train.select_dtypes(include=np.number), data.y.train)

results.update(
    grid,
    token=token,
    y_log=0,
    get_dummies=0,
    verbose=0,
).head(5)
```
